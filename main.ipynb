```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/train.csv")
df.head(15)
```

```python
df.isna().sum()
```

```python
df.info()
df.describe()
```

```python
def preprocessing(data):
    data = data.drop(['id'], axis=1)

    # data['f40_sqr'] = data['f40'] ** 2
    # data['f40_f70'] = data['f40'] * data['f70']

    data["row_std"] = data.std(axis=1)
    data["row_min"] = data.min(axis=1)
    # data["row_max"] = data.max(axis=1)
    # data["row_mean"] = data.mean(axis=1)
    data["row_sum"] = data.isna().sum(axis=1)

    return data
```

```python
from sklearn.model_selection import train_test_split


X = df.copy()
X = preprocessing(X)

y = X.pop('claim')

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.2, random_state=0
    )

X_train.head()
```

```python
from lightgbm import LGBMClassifier


model = LGBMClassifier(
    n_estimators=1000,
    random_state=0,
    learning_rate=0.05,
    num_leaves=64,
)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score


predict = model.predict_proba(X_valid)[:, 1]

features = model.feature_importances_
feature_names = X_train.columns

feat_importances = pd.Series(features, index=feature_names)
display(feat_importances.nlargest(20))


print(f"Accuracy: {roc_auc_score(y_valid, predict)}")
```

```python
test_df = pd.read_csv('data/test.csv')

test_X_df = preprocessing(test_df)

preds = model.predict_proba(test_X_df)[:, 1]

submission = pd.DataFrame({
    "id": test_df['id'],
    "claim": preds
})

submission.to_csv('submission.csv', index=False)
```
